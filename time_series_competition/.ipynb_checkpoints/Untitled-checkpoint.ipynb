```python
import pandas as pd 
import datetime
import numpy as np 
import matplotlib.pyplot as plt 
import statsmodels.api as sm
from pylab import rcParams
import warnings
from pandas.core.nanops import nanmean as pd_nanmean

from sklearn.metrics import mean_absolute_error

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df =  pd.read_csv('train.csv', parse_dates=['epoch'], index_col='epoch')
```

```python
df['minute'] = df.index.minute
df['hour'] = df.index.hour
df['day'] = df.index.day
```

```python
df['error']  = np.linalg.norm(df[['x', 'y', 'z']].values - df[['x_sim', 'y_sim', 'z_sim']].values, axis=1)
```

```python
sat_id_df = df[df['sat_id']==0]
sat_id_df[sat_id_df['day']==9]
```

```python
train_df = df[df['type']=='train']
test_df = df[df['type']=='test']
```

```python
plot_df = train_df[train_df['sat_id']==1]
```

```python
train_df_pivot = pd.pivot_table(plot_df, values = "error", columns='sat_id', index = "day")
```

```python
train_df_pivot.plot()
```

```python
fig, axes = plt.subplots(2, 1, sharey=False, sharex=False)
fig.set_figwidth(14)
fig.set_figheight(8)

window = 100
axes[0].plot(plot_df.index, plot_df.error, label='Original')
axes[0].plot(plot_df.index, plot_df.error.rolling(window=window).mean())
axes[0].legend(loc='best')
```
