```python
# 9
spec_df = df.groupby('spec')[['successful_orders', 'fail_orders']].sum().copy()
spec_df['fail_part'] = spec_df['fail_orders']/(spec_df['successful_orders'] + spec_df['fail_orders'])
spec_df.sort_values(by='fail_part', ascending=False).head(1)
```

```python
# 10
df_07 = df[('2019-07-01'<=df['date']) & (df['date']<'2019-08-01')]
df_07 = df_07.groupby('chain_id')['successful_orders'].sum().to_frame()

df_08 = df[('2019-08-01'<=df['date']) & (df['date']<'2019-09-01')]
df_08 = df_08.groupby('chain_id')['successful_orders'].sum().to_frame()

df_all = pd.merge(df_07, df_08, on='chain_id', how='inner')

df_all['abs_growth'] = df_all['successful_orders_y'] - df_all['successful_orders_x']
df_all.sort_values(by='abs_growth', ascending=False).head(1)
```
